```python
"""Here is a text adventure game about Popeye. One of the most famous cartoons
in the world.  His main rival is Bluto, his long time loving interest is Olive
Oil and his "Boy-Kid" is named Swee'pea. Bluto wants to get Olive Oil and is
always willing to capture her and Swee'pea but Popeye always finds his way out
with some spinach.

Here is how your stages are:
You will have 4 stages and an extra secret level. Once you accomplish the 
extra level you will finish the game.

Stage 1) Make a sailing decision.
Stage 2) What to do with Bluto?
Stage 3) What tool would you get for your performance?
Stage 4) Add the numbers.
Stage 5) Pick the right saying for Popeye. 

"""
import random

#This stage is makes you go to stage 2. You can fail. 
def stage_1():
    print("""
    Welcome to Popeye's game. You are Popeye, enjoying a beautiful sunset 
with your loved ones Olive and Swee'pea. Out of a sudden while you
were taking a shower, Bluto came in. He is your rival who is also in desperate
love for Olive, kidnapped both of them. Now, you are in your sailboat! You are
about to begin your mission.\n""")
    
    
    input(prompt = '<Press Enter to continue>\n')
    
    print("Enter 1 to start sailing and explore the area.")
    print("Enter 2 to remain at port.\n")
    
    choice = input(prompt = "> Your turn. Enter option 1 or 2\n")
    choice = choice.lower()
    
    #checking if player is entering 1, sail or start in their input. 
    if '1' in choice or 'sail' in choice or 'start' in choice:
        print("Get ready and start sailing.\n")
        stage_2() #You will be sent to stage 2.
        
   #checking if player is entering 2, remain or port in their input.
    elif '2' in choice or 'remain' or 'port' in choice:    
        print("""
    You are never going to find them if you don't go out to 
search for them.\n""")
        fail() #This sends you to the fail room.
        
    #other answers will make players retry current stage.
    else:
        print("Your entry was not valid.\n")
        input(prompt = "Press Enter to retry.\n")
        stage_1() #You will be sent to stage 1.

#This stage makes you decide what to do with Bluto. You can also fail and
#retry.
def stage_2():
    print("""
    You just found Bluto just 1 nautical mile north from your position and he 
has your "Boy-Kid" Swee'pea and your long time love interest Olive Oil. You 
really care of them, they are very special for you. What would you do Popeye?
\n""")
    
    input(prompt = '<Press Enter to continue>\n')
    
    print("Enter 1 to approach Bluto.")
    print("Enter 2 to remain quiet and wait for him.")
    print("Enter 3 to listen to music and drink some beer.\n")
    
    choice = input(prompt = "> Your turn. Enter option 1, 2, or 3.\n")
    choice = choice.lower()
    
    #checking if player is entering 1 or approach in their input.
    if '1' in choice or 'approach' in choice:
        print("You are right there, get closer.\n")
        stage_3() #You will be sent to stage 3.
        
    #checking if player is entering 2, remain or choice in their input.    
    elif '2' in choice or 'remain' in choice or 'wait' in choice:
        print("""
    As you decided to wait until an opportunity comes up.\n""")
        input(prompt = """After some time you decide to approach. Press Enter 
to retry.\n""")
        stage_2() #You will be sent to retry stage 2.
              
    #checking if player is entering 3, beer or music in their input.    
    elif '3' in choice or 'beer' in choice or 'music' in choice:
        print("""
    You have just distracted yourself quiet too much and now you
are unable to move. You just drank 25 beers.\n""")
        fail() #This sends you to the fail room.
              
    #other answers will make players retry current stage.    
    else:
        print("Your entry was not valid.\n")
        input(prompt = "Press Enter to retry.\n")
        stage_2() #You will be sent to retry stage 2.
        
    
#This stage makes you choose what tool would you get for Popeye's performance?
#You could also lose.
def stage_3():
    print("""
    You have come closer to him but he is really big and strong he just 
punched you and sent you off. Bluto is turning on the engines of his yacht to 
escape from you. Hurry up! You just found a few things in your sailboat. You
know exactly what you need from these picks before he departs. Focus on your
decision! What would you do?\n""")
    
    input(prompt = '<Press Enter to continue>\n')
    print("Enter 1 to take an anchor to defend yourself.")
    print("Enter 2 to eat a can of spinach.")
    print("Enter 3 to take a shower.\n")
    
    choice = input(prompt = "> Your turn. Enter option 1, 2, or 3.\n") 
    choice = choice.lower()
    
    #checking if player is entering 1, defend or anchor.          
    if '1' in choice or 'defend' in choice or 'anchor' in choice:
        print("""
    The anchor did not help at all as you tried to throw it at him but
he managed to catch it and threw it back to you, killing you.\n """)
        fail() #This sends you to the fail room.
              
    #checking if player is entering 2, eat or spinach.    
    elif '2' in choice or 'eat' in choice or 'spinach' in choice:
        print("""
    You have acquired maximum level of ultra strength and speed. You
defeated Bluto successfully and rescued your loved ones. You left bluto
unconscious and you guys escaped. A quick reminder, Spinach is your source of
power to beat Bluto. You become invincible.\n""")
        stage_4() #You will be sent to stage 4.
        
    #checking if player is entering 3 or shower.    
    elif '3' in choice or 'shower' in choice:
        print("""
    While showering, they escaped. Olive was screaming really loud
and you did not even listen to her. C'mon Popeye did you lose your mind? 
Why would you take a shower when you were that close to save them?\n""")
        fail() #This sends you to the fail room.

    #other answers will make players retry current stage.    
    else:
        print("Your entry was not valid.\n")
        input(prompt = "Press Enter to retry.\n")
        stage_3() #You will be sent to retry stage 3.

#This stage will claim a partial victory. It will take you to a final extra 
#game.      
def stage_4():
        
    #randomly generate 2 numbers.
    number1 = random.randint(15, 30)
    number2 = random.randint(15, 30)
    
    #This is relooping the stage 4 until you enter the right sum.     
    while True:
        print(f"""
    Bluto came back to port town of Sweethaven, you have to face him again.
You have already eaten even more spinach to reach unparalleled super human 
strength level. Bluto is already hurt and you have the option to win the 
battle and finally feel some freedom for a while. You have to sum {number1}
and {number2} and enter the right answer.\n""")
#inserting the 2 randomly generated numbers for the player to sum.
        
        popeyes_sum = input(prompt = '> Enter your answer here')
        
        #sum the two randomly generated numbers and convert them into a string.
        if popeyes_sum == str(number1 + number2):
            print("""
    You have made a super punch and he went Â¡Out of the Park! Your 
unparalleled strength level made you reach victory. Bluto is no longer around,
but there is one more little thing I need you to think about.\n""")#Make player feel suspicious about "little thing."
            secret_extralevel() #moves them to the next room.
            break
        #sum the two randomly generated numbers and convert them into a string.    
        elif popeyes_sum != str(number1 + number2): 
            
            print("Your entry was not correct.\n")
            input(prompt = "Press Enter to try again.\n")
        
        
        
#This stage will make you accomplish the game successfully. You can also fail.    
def secret_extralevel():
        print("""
    Now that the stress is all gone relax a little with this. Popeye, you are
particular with some of your phrases. Out of these phrases which one is yours?
\n""")
        
        input(prompt = '<Press Enter to continue>\n')
        print("Enter 1 for: It is what it is.")
        print("Enter 2 for: It's not rocket science.")
        print("Enter 3 for: I yam what I yam and that's all what I yam.\n")
        
        choice = input(prompt = "> Your turn. Enter option 1, 2, or 3.\n")
        choice = choice.lower()
        
        #checking if player is entering 1, it is what it is or it is.      
        if '1' in choice or 'it is what it is' in choice or 'it is' in choice:
            print("""That's not one of your phrases. You don't say that. C'mon
Popeye.\n""")
            fail() #This sends you to the fail room.
            
        #checking if player is entering 2, rocket, its not rocket science.
        elif '2' in choice or 'rocket' in choice or 'science' in choice:
            print("""That's not one of your phrases. You don't say that. C'mon
Popeye.\n""")
            fail() #This sends you to the fail room.
              
        #checking if player is entering 3, yam or I yam.    
        elif '3' in choice or 'yam' in choice or 'I yam' in choice:
            print("""
    That's your most popular phrase! It seems you know yourself more than you 
thought. You have just successfully passed the game. Thank you 
for playing dear Popeye.\n""")
            #Message for successfully beating the game.
        
             
def fail():
    print("You lose.")
    input(prompt = "Hit enter to exit.")
    
        
   
    
stage_1() # This starts the game 
```
